```python
import re

def split_into_sentences(text):
    """
    Splits text into sentences using punctuation.
    """
    return re.split(r'(?<=[.!?]) +', text.strip())
```

```python
from moviepy import VideoFileClip, CompositeVideoClip, TextClip, AudioFileClip, CompositeAudioClip
from gtts import gTTS
import random
import os

def create_subtitles_over_single_clip(video_path, story_text, output_path="final_video_with_subs.mp4"):
    sentences = split_into_sentences(story_text)
    
    video = VideoFileClip(video_path).without_audio()
    video_duration = video.duration  
    subtitle_clips = []
    tts_audio_clips = []
    current_time = 0
    
    for i, sentence in enumerate(sentences):
        audio_path = f"audio_sentence_{i}.mp3"
        tts = gTTS(sentence)
        tts.save(audio_path)
        audio = AudioFileClip(audio_path).with_start(current_time)
        print(audio_path, audio.duration)
        if current_time + audio.duration > video_duration:
            break

        text_clip = TextClip(
            text=sentence,
            font='Arial.ttf',
            font_size=24,
            color='white',
            text_align='center',
            size=(int(video.size[0] * 0.8), None),  
            method='caption'
        ).with_position(('center')).with_start(current_time).with_duration(audio.duration)

        subtitle_clips.append(text_clip)

        tts_audio_clips.append(audio)

        current_time += audio.duration  

        os.remove(audio_path)
        
    print(subtitle_clips, tts_audio_clips)
    combined_audio = CompositeAudioClip(tts_audio_clips)
    video = video.with_audio(combined_audio)
    final_video = CompositeVideoClip([video, *subtitle_clips])
    final_video.write_videofile(output_path, codec="libx264", audio_codec="aac")
```

```python
create_subtitles_over_single_clip(video_path='clips/output0.mp4', story_text=story, output_path="final_video_with_subs.mp4")
```
